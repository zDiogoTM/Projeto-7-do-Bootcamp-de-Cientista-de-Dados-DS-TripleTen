````python
df_weather = pd.read_csv('/datasets/project_sql_result_07.csv')
````

````python
df_saturdays = df_weather[df_weather['day_of_week'] == 'Saturday']

bad_weather = df_saturdays[df_saturdays['weather_conditions'] == 'Bad']
good_weather = df_saturdays[df_saturdays['weather_conditions'] == 'Good']

from scipy.stats import ttest_ind

bad_duration = bad_weather['duration_seconds'] / 60
good_duration = good_weather['duration_seconds'] / 60

t_stat, p_value = ttest_ind(bad_duration, good_duration, equal_var=False)

print(f'Estat√≠stica t: {t_stat:.3f}')
print(f'Valor-p: {p_value:.3f}')

alpha = 0.05
if p_value < alpha:
    print("Rejeitamos a hip√≥tese nula: h√° diferen√ßa significativa nas dura√ß√µes.")
else:
    print("Falha ao rejeitar a hip√≥tese nula: n√£o h√° evid√™ncia de diferen√ßa.")
````
